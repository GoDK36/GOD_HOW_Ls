```python
import os, re, csv
import pandas as pd
import numpy as np
from konlpy.tag import Mecab

mecab = Mecab(dicpath="C:\\mecab\\mecab-ko-dic")

quest_df = pd.read_csv(r"E:\대학교\HOWL's\챗봇 프로젝트\질문분류by10.csv", encoding='ANSI')

quest_df.shape
```

```python
noun_token = quest_df['명사토큰'].tolist()
quest_id = quest_df.iloc[:, 0].tolist()
big_class = ['이중', '부', '연계', '심화', '융합', '전공']
```

```python
quest_lst = [ [id, token] for id, token in zip(quest_id, noun_token)]
```

```python
classified_lst = []
for x in big_class:
    for id, token in quest_lst:
        if x in str(token):
            quest_df.iloc[id, 2] = x
            classified_lst.append(list(quest_df.iloc[id]))
```

```python
classified_df = pd.DataFrame(classified_lst)
classified_df.columns = ['id', '내용', '분류', '명사토큰']
```

```python
classified_df = classified_df.drop_duplicates('id', keep='first')
classified_df
```

```python
## 다른 파일 정리

eta_df = pd.read_csv(r"E:\Programming\python\NLP\HOW'Ls\Chatbot\질문.csv", encoding='ANSI')

eta_df = eta_df.drop(['글쓴이', 'id', '작성시간', '좋아요', '댓글 수'], axis=1)

eta_df
```

```python
eta_content = eta_df['내용'].tolist()
```

```python
a = [1, 2, 3, 4, 5,6]
del a[1:]
a
```

```python
from kiwipiepy import Kiwi

kiwi = Kiwi()

#대분류 사전 추가
kiwi.add_user_word('이중전공', 'DM', 3.0)
kiwi.add_user_word('융합전공', 'UM', 3.0)
kiwi.add_user_word('전공심화', 'MD', 3.0)
kiwi.add_user_word('부전공', 'SM', 3.0)
kiwi.add_user_word('연계전공', 'UM', 3.0)

kiwi.prepare()

lem_cont = []
res_cont = []
for question in eta_content[:5]:
    for res, scr in kiwi.analyze(question, top_n=1):
        for result in res:
            res_cont.append(list(result))
        lem_cont.append(res_cont)
        
# print(lem_cont)
```

```python
merge_lem = []

for morph in lem_cont:
    txt = ' '.join(morph)
    merge_lem.append(txt)

print(merge_lem[:10])

eta_df['명사토큰'] = merge_lem

eta_df
```

```python
eta_id = quest_df.iloc[:, 0].tolist()
big_class = ['이중전공', '부전공', '연계전공', '전공심화', '융합전공']

eta_quest_lst = [ [id, token] for id, token in zip(eta_id, lem_cont)]
```

```python
eta_classified_lst = []
for x in big_class:
    for id, token in eta_quest_lst:
        if x in str(token):
            eta_df.iloc[id, 3] = x
            eta_classified_lst.append(list(eta_df.iloc[id]))
```
